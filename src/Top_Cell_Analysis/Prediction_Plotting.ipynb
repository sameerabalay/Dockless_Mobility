```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import sys, os
```

```python
predictions_df = pd.read_pickle("../../data/pickled_data/test_predictions.pkl")
```

```python
predictions_df.head(5)
```

```python
pred_rf = predictions_df[['Actual', 'rf_predictions']]
pred_ar = predictions_df[['Actual', 'ar_predictions']]
pred_ma = predictions_df[['Actual', 'ma_predictions']]
```

```python

pred_rf.plot(style=['-', '--', ':'],figsize=(100, 5))
plt.title('Predictions Random Forest')
plt.ylabel('Hourly scooter count', fontsize=12)
plt.savefig('../../output/Random_Forest_Predictions.png')

pred_ar.plot(style=['-', '--', ':'],figsize=(100, 5))
plt.title('Predictions AR Model')
plt.ylabel('Hourly scooter count', fontsize=12)
plt.savefig('../../output/AR_Predictions.png')

pred_ma.plot(style=['-', '--', ':'],figsize=(100, 5))
plt.title('Predictions MA Model')
plt.ylabel('Hourly scooter count', fontsize=12)
plt.savefig('../../output/MA_Predictions.png')


plt.show()
```
